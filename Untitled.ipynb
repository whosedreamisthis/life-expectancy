```python
import pandas as pd
import joblib
# from le_package import config

def test_prediction():
    # 1. Load the Data
    # config.DATA_DIR and config.TRAINING_DATA_FILE were defined in your config.py
    df = pd.read_csv("data/Life Expectancy Data.csv")
    df.columns = df.columns.str.strip()
    df.dropna(subset=["Life expectancy"], inplace=True)
    # df = data.sort_values(['Country', 'Year'])

    X = df.drop("Life expectancy", axis=1)
    y = df["Life expectancy"]
    
    # 3. Load the Saved Model
    # Construct the path to your specific pkl file
    
    
    
    pipeline = joblib.load("models/le_model_v0.1.0.pkl")
    
    # 4. Make Predictions
    predictions = pipeline.predict(X)
    
    # 5. Evaluate
    score = pipeline.score(X, y)
    print(f"Model R2 Score on full dataset: {score:.4f}")
    
    # Show some sample results
    results = pd.DataFrame({
        'Actual': y,
        'Predicted': predictions
    })
    print("\nSample Predictions:")
    print(results.head())

if __name__ == "__main__":
    test_prediction()
```
