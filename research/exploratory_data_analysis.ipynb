```python
# Sort by country and year first to ensure interpolation makes sense
data = data.sort_values(['Country', 'Year'])

# Fill missing values using linear interpolation within each country
num_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[num_cols] = data.groupby('Country')[num_cols].transform(lambda x: x.interpolate(limit_direction='both'))

data['GDP'] = data['GDP'].fillna(data.groupby('Status')['GDP'].transform('median'))
data['Population'] = data['Population'].fillna(data.groupby('Status')['Population'].transform('median'))

data[num_cols] = data[num_cols].fillna(data[num_cols].median())
```

```python
# 2. Define the grid layout (4 columns wide)
n_cols = 4
n_rows = (len(num_vars) + n_cols - 1) // n_cols

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20, 5 * n_rows))
axes = axes.flatten()

# 3. Loop and plot
for i, col in enumerate(num_vars):
    sns.boxplot(y=data[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)
    axes[i].set_ylabel('')

# 4. Remove empty slots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import country_converter as coco
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Initialize the converter
cc = coco.CountryConverter()

# 2. Convert country names to continents
# .unique() makes it faster, then we map it back to the main dataframe
unique_countries = data['Country'].unique()
continent_map = cc.convert(names=unique_countries, to='continent')
country_to_continent = dict(zip(unique_countries, continent_map))

data['Continent'] = data['Country'].map(country_to_continent)
plt.figure(figsize=(10, 6))

# Update: Assign 'Continent' to both x and hue
sns.boxplot(
    data=data, 
    x='Continent', 
    y='Life expectancy', 
    hue='Continent',  # This matches the x variable
    palette='Set2', 
    legend=False      # Removes the redundant legend
)

plt.title('Life Expectancy Distribution by Continent')
plt.xticks(rotation=45)
plt.show()

# 4. Check for missing values in the conversion
print(f"Countries not matched: {data[data['Continent'] == 'not found']['Country'].unique()}")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Select the immunization columns (strip names to be safe)
imm_cols = ['Hepatitis B', 'Polio', 'Diphtheria']
imm_df = data[imm_cols].dropna()

# 2. Calculate VIF
vif_data = pd.DataFrame()
vif_data["feature"] = imm_df.columns
vif_data["VIF"] = [variance_inflation_factor(imm_df.values, i) for i in range(len(imm_df.columns))]

print(vif_data)
```

```python
# Create the average score across the three vaccines
data['Immunization_Score'] = data[['Hepatitis B', 'Polio', 'Diphtheria']].mean(axis=1)

# Check the new correlation
new_corr = data[['Immunization_Score', 'Life expectancy']].corr().iloc[0,1]
print(f"New Composite Correlation: {new_corr:.3f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compare the relationship of each vaccine vs the composite score
cols_to_plot = ['Hepatitis B', 'Polio', 'Diphtheria', 'Immunization_Score']

fig, axes = plt.subplots(1, 4, figsize=(20, 5), sharey=True)
for i, col in enumerate(cols_to_plot):
    sns.regplot(data=data, x=col, y='Life expectancy', ax=axes[i], 
                scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
    axes[i].set_title(f'{col} vs Life Expectancy')

plt.tight_layout()
plt.show()
```
